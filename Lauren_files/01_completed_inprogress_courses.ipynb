```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
plt.style.use('ggplot')

%matplotlib inline
```

```python
survey_2018 = pd.read_csv('./resources/04_Kaggle_Survey_2018.csv')
survey_2018 = survey_2018.drop([0],axis=0)
survey_2018.head(2)
```

```python
total_2018 = survey_2018['Time from Start to Finish (seconds)'].count()
total_2018
```

```python
courses_2018 = survey_2018[['Q36_Part_1','Q36_Part_2','Q36_Part_3','Q36_Part_4','Q36_Part_5','Q36_Part_6'
                                    ,'Q36_Part_7','Q36_Part_8','Q36_Part_9','Q36_Part_10','Q36_Part_11','Q36_Part_12'
                                    ,'Q36_Part_13','Q36_OTHER_TEXT']]
courses_2018
```

```python
udacity = (courses_2018.count()["Q36_Part_1"])
coursera = courses_2018.count()["Q36_Part_2"]
edX = courses_2018.count()["Q36_Part_3"]
data_camp = courses_2018.count()["Q36_Part_4"]
data_quest = courses_2018.count()["Q36_Part_5"]
kaggle_learn = courses_2018.count()["Q36_Part_6"]
fast_ai = courses_2018.count()["Q36_Part_7"]
developers_google = courses_2018.count()["Q36_Part_8"]
udemy = courses_2018.count()["Q36_Part_9"]
the_schoolai = courses_2018.count()["Q36_Part_10"]
online_university = courses_2018.count()["Q36_Part_11"]
none = courses_2018.count()["Q36_Part_12"]
other = courses_2018.count()["Q36_Part_13"]
```

```python
udacity = (udacity / total_2018) * 100 
coursera = (coursera / total_2018) * 100 
edX = (edX / total_2018) * 100 
data_camp = (data_camp / total_2018) * 100 
data_quest = (data_quest / total_2018) * 100 
kaggle_learn = (kaggle_learn / total_2018) * 100 
fast_ai = (fast_ai / total_2018) * 100 
developers_google = (developers_google / total_2018) * 100 
udemy = (udemy / total_2018) * 100 
the_schoolai = (the_schoolai / total_2018) * 100 
online_university = (online_university / total_2018) * 100 
none = (none / total_2018) * 100 
other = (other / total_2018) * 100 
```

```python
df = pd.DataFrame({'Courses':['Udacity', 'Coursera', 'edX','DataCamp','DataQuest','Kaggle Learn','fast.ai',
                                  'Developers.google.com','Udemy','The.School.ai','Online University','None',
                                  'Other'], 'Responded Yes': [udacity, coursera, edX,data_camp,data_quest,kaggle_learn,
                                   fast_ai,developers_google,udemy,the_schoolai,online_university,none,other]})
df = df.sort_values('Responded Yes')

df
```

```python
ax = df[['Responded Yes']].plot(kind='barh',
              figsize=(10,7), color=['dodgerblue', 'slategray'], fontsize=13);

ax.set_alpha(.7)
ax.set_title("2018 Completed / InProgress Courses",fontsize=18)
ax.set_xlabel("Total Responses", fontsize=18)
ax.set_ylabel("Courses", fontsize=18)
ax.set_xticks([0, 10, 20, 30, 40, 50])
ax.set_yticklabels(df['Courses'])

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+1, i.get_y()+.20, \
            str(round((i.get_width()), 2))+'%', fontsize=11, color='dimgrey')

fig = ax.get_figure()

# invert for largest on top 
#ax.invert_yaxis()
ax.legend(loc='lower right')
plt.show()
fig.tight_layout()
fig.savefig("completed_inprogress_courses.png")
```

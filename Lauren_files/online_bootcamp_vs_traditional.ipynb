```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
plt.style.use('ggplot')
%matplotlib inline
```

```python
survey_2018 = pd.read_csv('./resources/04_Kaggle_Survey_2018.csv')
survey_2018 = survey_2018.drop([0],axis=0)
survey_2018.head(2)
```

```python
total_2018 = survey_2018['Time from Start to Finish (seconds)'].count()
total_2018
```

```python
df_39 = survey_2018[['Q39_Part_1']]
df_39.head()
```

```python
slightly_better = df_39.loc[df_39["Q39_Part_1"] == "Slightly better"].count()
no_opinion = df_39.loc[df_39["Q39_Part_1"] == "No opinion; I do not know"].count()
much_better = df_39.loc[df_39["Q39_Part_1"] == "Much better"].count()
neither = df_39.loc[df_39["Q39_Part_1"] == "Neither better nor worse"].count()
slightly_worse = df_39.loc[df_39["Q39_Part_1"] == "Slightly worse"].count()
much_worse = df_39.loc[df_39["Q39_Part_1"] == "Much worse"].count()

slightly_better = (slightly_better / total_2018)* 100 
no_opinion = (no_opinion / total_2018)* 100 
much_better = (much_better / total_2018)* 100 
neither = (neither / total_2018)* 100 
slightly_worse = (slightly_worse / total_2018)* 100 
much_worse = (much_worse / total_2018)* 100 
```

```python
responses_39 = pd.DataFrame({
                        'Responses':["Much Better", "Slightly Better",'Neither Better nor Worse',"Slightly Worse","Much Worse","No Opinion"], 
                          'Totals':[much_better, slightly_better, neither, slightly_worse, much_worse, no_opinion]})
responses_39
```

```python
responses_39 = responses_39.astype({'Responses': object, 'Totals': int})
responses_39.dtypes
```

```python
responses_39
```

```python
df_39_2 = survey_2018[['Q39_Part_2']]
df_39_2.head()
```

```python
slightly_better_2 = df_39_2.loc[df_39_2["Q39_Part_2"] == "Slightly better"].count()
no_opinion_2 = df_39_2.loc[df_39_2["Q39_Part_2"] == "No opinion; I do not know"].count()
much_better_2 = df_39_2.loc[df_39_2["Q39_Part_2"] == "Much better"].count()
neither_2 = df_39_2.loc[df_39_2["Q39_Part_2"] == "Neither better nor worse"].count()
slightly_worse_2 = df_39_2.loc[df_39_2["Q39_Part_2"] == "Slightly worse"].count()
much_worse_2 = df_39_2.loc[df_39_2["Q39_Part_2"] == "Much worse"].count()

slightly_better_2 = (slightly_better_2 / total_2018)* 100 
no_opinion_2 = (no_opinion_2 / total_2018)* 100 
much_better_2 = (much_better_2 / total_2018)* 100 
neither_2 = (neither_2 / total_2018)* 100 
slightly_worse_2 = (slightly_worse_2 / total_2018)* 100 
much_worse_2 = (much_worse_2 / total_2018)* 100 
```

```python
df_39_2 = pd.DataFrame({'Responses':["Much Better", 
                                "Slightly Better",
                                'Neither Better nor Worse',
                                "Slightly Worse",
                                "Much Worse",
                                "No Opinion"], 
                          'Totals':[much_better_2, slightly_better_2, neither_2, slightly_worse_2, much_worse_2, no_opinion_2]})

df_39_2
```

```python
df_39_2 = df_39_2.astype({'Responses': object, 'Totals': int})
df_39_2
```

```python
merged_39 = pd.merge(responses_39, df_39_2 ,on="Responses",how="left")
merged_39.columns = ['Responses','online','bootcamp']
merged_39
```

```python
ax = merged_39[['online','bootcamp']].plot(kind='barh',
              figsize=(10,7), color=['dodgerblue', 'slategray'], fontsize=13);
ax.set_alpha(.7)
ax.set_title('Online Courses / BootCamps VS. Traditional Education',fontsize=18)
ax.set_xlabel("Total Responses", fontsize=18)
ax.set_ylabel("Perceived Quality", fontsize=18)
ax.set_xticks([0, 5, 10, 15, 20, 25,30])
ax.set_yticklabels(merged_39['Responses'])

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width() + .5, i.get_y() +.155, \
            str(round((i.get_width()), 2))+'%', fontsize=11, color='dimgrey')
fig = ax.get_figure()

# invert for largest on top 
ax.invert_yaxis()
ax.legend(loc='lower right')
plt.show()
fig.tight_layout()
fig.savefig("online_bootcamp_vs_trad.png")
```
